```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (12,8) # set figure size
```

# Law of large number
* The following code to show the convergence of mean of x, when x follows N(5, 1)

```python
mu = 5
nsim = 100
N_all = [10, 100, 1000, 10000]
OUT_all = np.zeros((nsim, len(N_all)))

for i in range(nsim):
    for j in range(len(N_all)):
        OUT_all[i, j] = np.mean(np.random.normal(mu, 1, size=N_all[j]))

for j in range(len(N_all)):
    plt.subplot(2, 2, j+1)
    plt.hist(OUT_all[:, j])
    plt.xlim(3.8, 6.2)
    plt.ylim(0, 30)
    plt.ylabel('Frequency')
    plt.title('N='+str(N_all[j]))
```

# Central limit theorem 
* The following code to show asymptotic distribution of mean of x, when x follows Bernoulli(0.1)

```python
n_all = [1, 10, 100, 10000]
x = np.random.binomial(1000*10000*[1], 0.1).reshape(1000, 10000) # row = 1000, col = 10000

for i in range(len(n_all)):
    n = n_all[i]
    x_bar = np.mean(x[:, 0:n], axis = 1)
    z = (x_bar - 0.1)*np.sqrt(n) / np.sqrt(0.1*0.9)
    plt.subplot(2, 2, i+1)
    plt.hist(z)
    plt.ylabel('Frequency')
    plt.title('n='+str(n))
```
